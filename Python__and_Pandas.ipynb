# Leitura e informaes detalhadas sobre o arquivo
O Pandas tem algumas funes para ler e converter dados: read_csv, read_excel e read_table que so utilizados com mais frequncia. 
No nosso caso  feito a leitura do arquivo com o read_csv e depois usamos o head() para mostrar as 5 primeiras linhas do nosso arquivo, podemos usar head(10) para mostrar as 10 primeiras linhas ou um outro valor qualquer de sua preferncia.

```python
df = pd.read_csv('https://raw.githubusercontent.com/GabrielCheu/Learn_pandas_with_Pokemon/main/pokemon_data.csv')
df.head()
```

# **Consultando dados com o loc**
Com o mtodo loc podemos selecionar um subconjunto de linhas e colunas de um DataFrame usando a indexao baseada em um rtulo (ttulo das colunas).

```python
df.loc[[1,2,5]]
```

O loc tambm  usado para fazer o fatiamento, a sintaxe  df.loc[start : stop],

```python
df.loc[20:25]
```

Podemos pegar as lisnhas correspondentes a uma coluna especifica e ainda caso queira especificar qual dado da coluna queira mostrar.

```python
df.loc[df['Generation']==3]
```

```python
df.loc[df['Type 1'] == 'Dragon', ["#","Name","Attack"]].head()
```

```python
df.groupby('Type 1').sum()
```

```python
#criando pokemon Cheumander
new_pokemon = {'#':['800'], 'Name':['Cheumander'], 'Type 1':['Dragon'],'Attack':['200'], 'Legendary':['1']}
poke_df = pd.DataFrame(new_pokemon)
poke_df.head()
```

```python
#adicionando linha
df = df.append(poke_df)
df.tail()
```

```python
#deletando linha
df = df.drop(0, axis=0)
df.tail()
```
