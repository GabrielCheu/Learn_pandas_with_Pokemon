# Importando a biblioteca do pandas
Primeiro e mais importante é chamar a bibliote do pandas com o comando: import pandas

Em seguida para deixar seu codigo de programação mais limpo e fluido, de um apelido para sua biblioteca, para facilitar na hora de chamar ela, na maioria esmagadora das vezes os programadores chamam o pandas de pd.

```python
#bibliotecas
import pandas as pd
```

# Leitura e informações detalhadas sobre o arquivo
O Pandas tem algumas funções para ler e converter dados: read_csv, read_excel e read_table que são utilizados com mais frequência. 
No nosso caso é feito a leitura do arquivo com o read_csv e depois usamos o head() para mostrar as 5 primeiras linhas do nosso arquivo, podemos usar head(10) para mostrar as 10 primeiras linhas ou um outro valor qualquer de sua preferência.

```python
df = pd.read_csv('https://raw.githubusercontent.com/GabrielCheu/Learn_pandas_with_Pokemon/main/pokemon_data.csv')
df.head()
```

Da mesma maneira podemos utilizar o tail() para ver as 5 últimas linhas.

```python
df.tail()
```

O comando shape fornece as dimensões do DataFrame

```python
df.shape
```

O df.columns retorna uma lista com o nome de todas as colunas do DataFrame.

```python
df.columns
```

# **Analise Geral de dados**

A função describe() retorna o resumo estatístico do dataframe. Isso inclui a variação padrão de contagem, média, mediana, valores mínimos e máximos das colunas.

```python
df.describe()
```

```python
atk = df[['#', 'Name', 'Attack']]
atk.head()
```

```python
atk = df[['#', 'Name', 'Attack']].sort_values(by='Attack', ascending=False)
atk.head()
```

```python
df.groupby('Type 1').max()
```

# **Consultando dados com o loc**
Com o método loc podemos selecionar um subconjunto de linhas e colunas de um DataFrame usando a indexação baseada em um rótulo (título das colunas).

```python
df.loc[[1,2,5]]
```

O loc também é usado para fazer o fatiamento, a sintaxe é df.loc[start : stop],

```python
df.loc[20:25]
```

Podemos pegar as lisnhas correspondentes a uma coluna especifica e ainda caso queira especificar qual dado da coluna queira mostrar.

```python
df.loc[df['Generation']==3]
```

```python
df.loc[df['Type 1'] == 'Dragon', ["#","Name","Attack"]].head()
```

```python
df.groupby('Type 1').sum()
```

```python
#criando pokemon Cheumander
new_pokemon = {'#':['800'], 'Name':['Cheumander'], 'Type 1':['Dragon'],'Attack':['200'], 'Legendary':['1']}
poke_df = pd.DataFrame(new_pokemon)
poke_df.head()
```

```python
#adicionando linha
df = df.append(poke_df)
df.tail()
```

```python
#deletando linha
df = df.drop(0, axis=0)
df.tail()
```
